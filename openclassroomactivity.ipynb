```
#verification de notre dataset ainsi transformé
naissances.head()
```

Afin de pouvoir faire quelques manipulations, nous avons besoin de mettre notre index sur 'date' et ainsi transformer notre dataframe  en série temporelle. Ce qui est bien pratique pour manipuler nos données à partir de la date.

```
#mettons notre index sur 'date'
naissances= naissances.set_index('date')
```

```
naissances.head()
```

```
#regardons nos données pour tous les 2 du mois en utilisant 'index.day'
naissances[naissances.index.day == 2]
```

```
#regardons le nombre de naissances pour une date précise
naissances.loc['1988-01-02','births']


```

```
#selectionnons les données  entre certaines dates
naissances.loc['1988-01-02': '1988-03-02']
```

```
#calculons la moyenne des naissances par jour en utilisant la methode resample
#cette methode permet de réechantilloner nos données :ici nous voulons les moyennes par mois 'M' mais il est possible de les avoir par jour ('D') ou par année ('Y')
naissances.resample('M').mean().head()
```

```
#nous pouvons prendre le min, le max, la sum....

naissances.resample('M').max().head()
```

```
naissances.dtypes
```

```
# calculons les moyennes glissantes sur 7 periodes
rolling_mean =naissances['births'].rolling(7,center=True).mean()
```

```
naissances['rolling_mean']= naissances['births'].rolling(7,center=True).mean()
```

```
naissances.head(10)
```

```
#créons une nouvelle colonne 'births'qui englobe les naissances filles et gaçons
naissances.resample('D').sum()
```
